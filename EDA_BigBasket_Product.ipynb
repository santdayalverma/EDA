__Top & least sold products__

```python
top_Product=df['product'].value_counts().head(10)
least_Product=df['product'].value_counts().tail(10)
```

```python
# Calculate the top product counts
top_product = df['product'].value_counts().head(10)

# Set the maximum width for wrapping the x-axis labels
max_width = 10

# Apply text wrapping to the x-axis labels
wrapped_labels = [textwrap.fill(label, width=max_width) for label in top_product.index]

# Create a bar plot of the top products with wrapped x-axis labels
plt.figure(figsize=(14, 5))
plt.bar(wrapped_labels, top_product.values)
plt.xlabel('Product')
plt.ylabel('Count')
plt.title('Top Products')
#plt.xticks(ha='right')  # Rotate and align the x-axis labels
plt.tight_layout()  # Adjust the layout to prevent overlapping labels
plt.show()
```

```python
# Calculate the least product counts
top_product = df['product'].value_counts().tail(10)

# Set the maximum width for wrapping the x-axis labels
max_width = 10

# Apply text wrapping to the x-axis labels
wrapped_labels = [textwrap.fill(label, width=max_width) for label in top_product.index]

# Create a bar plot of the least products with wrapped x-axis labels
plt.figure(figsize=(14, 5))
plt.bar(wrapped_labels, top_product.values)
plt.xlabel('Product')
plt.ylabel('Count')
plt.title('least Products')
#plt.xticks(rotation=90 ,ha='right')  # Rotate and align the x-axis labels
plt.tight_layout()  # Adjust the layout to prevent overlapping labels
plt.show()
```

**Through the plots of most and least sold products, we get to know that**

1. Indians have a trend to purchase different types of oils and ghee.
2. Turmeric and coriander are the most used spices in indian cuisine. On the other hand, Nutmeg and pepper powder are used in scarce amounts.
3. Hand santizer is one of the highest sold products most probably due to covid-19 pandemic.
4. Indians are very keen about haircare as we see hair color, shampoo and oils in the highest selling products.
5. Raw spices like turmeric are preferred on masala mixtures.

Effect of discounted prices

```python
df['Diff_Price']=df['market_price']-df['sale_price']
Discount_Prodct_Sales=df[df['Diff_Price'] !=0]
Discount_Prodct_Sales
```

```python
sns.boxplot(Discount_Prodct_Sales['market_price'], showmeans=True, color='r')
```

```python
#sns.boxplot(Discount_Prodct_Sales['market_price'], showmeans=True)
```

```python
fig = plt.figure(figsize=(15, 7))

plt.style.use("Solarize_Light2")
sns.distplot(df.rating, color='b', kde=True, label='All Products')
sns.distplot(Discount_Prodct_Sales.rating, color='gold', kde=True, label='Discounted Products')
plt.xlabel("Ratings", fontsize=15, weight='semibold')
plt.ylabel("Density", fontsize=15, weight='semibold')
plt.title("Relative distribution of all products with discounted products", fontsize=15, weight='semibold')
plt.legend()
plt.show()
```

```python
fig = plt.figure(figsize=(20,10))

plt.style.use("Solarize_Light2")
sns.distplot(Discount_Prodct_Sales.rating, color='b', kde =True)
sns.distplot(df.rating, color='gold', kde =True)
plt.xlabel("Ratings",fontsize=15, weight='semibold')
plt.ylabel("Density",fontsize=15, weight='semibold')
plt.title("Relative distribution of all products with discounted products",fontsize=15, weight='semibold')
fig.legend()
```
